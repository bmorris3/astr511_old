```python
%pylab inline
import numpy as np
matplotlib.rcParams['font.size'] = 16
from matplotlib import pyplot as plt
import pyfits
from astroML import datasets
```

### Load stars from astroML

```python
# Load data from Ivezic et al. 2007
stripe82raw = datasets.fetch_sdss_S82standards(data_home=None, download_if_missing=True, crossmatch_2mass=True)[::1000]
bands = ['u', 'g', 'r', 'i', 'z', 'J', 'H', 'K']
stripe82bands = ['mmed_u', 'mmed_g', 'mmed_r', 'mmed_i', 'mmed_z', 'J', 'H', 'K']
s = {}
for band, stripe82band in zip(bands, stripe82bands):
    s[band] = stripe82raw[stripe82band]
```

### Load quasars from astroML

```python
quasars = datasets.fetch_dr7_quasar()[::100]
quasarbands = ['mag_u', 'mag_g', 'mag_r', 'mag_i', 'mag_z', 'mag_J', 'mag_H', 'mag_K']
q = {}
for band, quasarband in zip(bands, quasarbands):
    q[band] = quasars[quasarband]
```

### Load galaxies from [NYU VAGC](http://sdss.physics.nyu.edu/vagc/)

```python
sdss = pyfits.open('object_sdss_imaging.fits')[1].data[::1e3]
twomass = pyfits.open('object_twomass.fits')[1].data[::1e3]

# Only use data from galaxies with non-zero 2MASS fluxes
nonzeromags = (twomass['J_M_K20FE'] != 0)*(twomass['H_M_K20FE'] != 0)*(twomass['K_M_K20FE'] != 0)
galaxies = np.zeros((8, sum(condition)))
galaxies[5, :] = twomass['J_M_K20FE'][nonzeromags]
galaxies[6, :] = twomass['H_M_K20FE'][nonzeromags]
galaxies[7, :] = twomass['K_M_K20FE'][nonzeromags]

# Convert nMg to magnitudes for SDSS
SDSSnanomaggies = sdss['PSFFLUX']
SDSSmags = 22.5 - 2.5*np.log10(SDSSnanomaggies)

# Make big matrix of fluxes
galaxies[:5, :] = SDSSmags[nonzeromags].T

g = {}
for i, band in enumerate(bands):
    g[band] = galaxies[i]
```

### Plot the objects

```python
def plotcolors(axis, color1, color2, color3, color4, xlim, ylim):
    '''
    Plot on x-axis `color1` - `color2`, and `color3` - `color4` on y-axis. 
    Use formatting string and set limits to x, y axes with `xlim`, `ylim`
    '''
    axis.plot(s[color1] - s[color2], s[color3] - s[color4], 'r.', label='Stars')
    axis.plot(q[color1] - q[color2], q[color3] - q[color4], 'b.', label='QSOs')
    axis.plot(g[color1] - g[color2], g[color3] - g[color4], 'm.', label='Galaxies')
    axis.set_xlabel("{0} - {1}".format(color1, color2))
    axis.set_ylabel("{0} - {1}".format(color3, color4))
    axis.set_xlim(xlim)
    axis.set_ylim(ylim)

fig, ax = plt.subplots(2, 2, figsize=(16, 14))

plotcolors(ax[0, 0], 'g', 'r', 'r', 'i', [-0.5, 2], [-0.5, 2])
plotcolors(ax[0, 1], 'g', 'i', 'J', 'K', [-1, 4], [0, 2.5])
plotcolors(ax[1, 0], 'J', 'H', 'H', 'K', [0, 1], [0, 1])
plotcolors(ax[1, 1], 'g', 'i', 'i', 'K', [-1, 4], [0, 5])

# Annotate upper right plot
ax[0, 1].annotate('QSOs', (0, 2.3), textcoords='data', fontsize=18, color='b')
ax[0, 1].annotate('Galaxies', (1.5, 1.7), textcoords='data', fontsize=18, color='m')
ax[0, 1].annotate('Stars', (2, 0.2), textcoords='data', fontsize=18, color='r')

# Annotate lower right plot
ax[1, 1].annotate('QSOs', (-0.5, 4.5), textcoords='data', fontsize=18, color='b')
ax[1, 1].annotate('Galaxies', (2, 4), textcoords='data', fontsize=18, color='m')
ax[1, 1].annotate('Stars', (2, 1.5), textcoords='data', fontsize=18, color='r')

# Plot classification boundaries
plotline = lambda xmin, xmax, slope, intercept: (np.linspace(xmin, xmax), np.linspace(xmin, xmax)*slope + intercept)
lineprops = {'color':'k', 'ls':'--', 'lw':2}
ax[0, 1].plot(*plotline(0, 4, 0.25, 0.5), **lineprops)
ax[0, 1].plot(*plotline(0, 2, 1, 0.5), **lineprops)
ax[1, 1].plot(*plotline(0, 4, 0.8, 1.4), **lineprops)
ax[1, 1].axvline(x=0.9, ymin=0.45, ymax=1, **lineprops)

ax[0, 0].legend(loc='upper left', numpoints=1)

plt.show()
```
